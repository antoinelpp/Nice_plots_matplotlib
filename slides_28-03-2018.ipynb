# How to represent data
**Using Python**

```python
import matplotlib.pyplot as plt
savepath = "./LaTex/images/"
```

```python
def plot():
    plt.plot([1,2,3,4,5,6,7,8,9],[3,4,2,2,3,7,5,6,4])
    plt.xlabel(r"$ \omega$")
    plt.ylabel(r"E [$cm$]")
    plt.title(r"this this the Title")
```

```python
#plt.style.use("seaborn")
plot()
plt.savefig(savepath+"default")
```

```python
with plt.style.context(('presentation')):
    plot()
plt.savefig(savepath+"custom")
```
